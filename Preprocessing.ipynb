> First read in PDB14189 and PDB2272 in .txt format. Then we do some simple preprocessing to combine them into csv files.

```python
with open('PDB2272_N.txt') as f:
  test_N = f.readlines()
with open('PDB2272_P.txt') as f:
  test_P = f.readlines()
with open('PDB14189_N.txt') as f:
  train_N = f.readlines()
with open('PDB14189_P.txt') as f:
  train_P = f.readlines()
```

```python
test_N = [test_N[i] for i in range(1, len(test_N), 2)]
test_P = [test_P[i] for i in range(1, len(test_P), 2)]
train_N = [train_N[i] for i in range(1, len(train_N), 2)]
train_P = [train_P[i] for i in range(1, len(train_P), 2)]
```

```python
X_test = test_N + test_P
X_train = train_N + train_P
y_test = [0] * len(test_N) + [1] * len(test_P)
y_train = [0] * len(train_N) + [1] * len(train_P)
```

```python
test = list(zip(X_test, y_test))
train = list(zip(X_train, y_train))
train = pd.DataFrame(train, columns = ['ProteinSequence', 'Class'])
test = pd.DataFrame(test, columns = ['ProteinSequence', 'Class'])
```

```python
# randomly shuffle the csvs and save them
train_data=train.sample(frac=1).reset_index(drop=True)
test_data=test.sample(frac=1).reset_index(drop=True)

train_data.to_csv('14189_shuffled.csv') # this is PDB14189
test_data.to_csv('2272_shuffled.csv') # this is PDB2272 
```

> Then we read in PDB1075 and PDB186 and shuffle them.

```python
train = pd.read_csv("PDB1075.csv")
test = pd.read_csv("testData186.csv")

train_data=train.sample(frac=1, random_state=42).reset_index(drop=True)
test_data=test.sample(frac=1, random_state=42).reset_index(drop=True)

train_data.to_csv('1075_shuffled.csv')
test_data.to_csv('186_shuffled.csv')
```
